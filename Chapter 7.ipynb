```python
x = np.array([1, 2, 3, 4])
y = np.array([1, 3, 5, 4])
deg = len(x) - 1
A = P.polynomial.polyvander(x, deg)
c = linalg.solve(A, y)
f1=P.Polynomial(c)
```

```python
f1
```

```python
xx = np.linspace(-5,5)
plt.scatter(x, y)
plt.plot(xx,f1(xx))
```
